
#### 1️⃣ Joint Probability 

####  What is Joint Probability?

Joint probability is the **probability of two events occurring at the same time**.

> **Notation:**
> $P(A \cap B)$ or $P(A, B)$
> "The probability that **both A and B** happen."

---

####  Example

> Suppose:
>
> * A = event of rolling an even number on a die
> * B = event of rolling a number > 3

* Possible outcomes = {1, 2, 3, 4, 5, 6}
* A = {2, 4, 6}
* B = {4, 5, 6}
* $A \cap B = \{4, 6\}$

So,
**Joint probability:**

$$
P(A \cap B) = \frac{\text{Number of outcomes in } A \cap B}{\text{Total outcomes}} = \frac{2}{6} = \frac{1}{3}
$$

---

####  General Formula

#### ✅ If A and B are **independent**:

$$
P(A \cap B) = P(A) \cdot P(B)
$$

#### ❌ If A and B are **dependent**:

$$
P(A \cap B) = P(A) \cdot P(B|A)
$$

---

####  Real-Life Example

> Medical Test:

* A = Person has the disease (5% of people)
* B = Test shows positive result
* $P(A) = 0.05$, $P(B|A) = 0.9$ (test is 90% accurate)

So:

$$
P(A \cap B) = P(A) \cdot P(B|A) = 0.05 \cdot 0.9 = 0.045
$$

→ There's a 4.5% chance a person **has the disease and tests positive**.




````python

# Joint probability from simulated data
import pandas as pd

data = pd.DataFrame({
    'A': ['Yes', 'No', 'Yes', 'Yes', 'No', 'No', 'Yes', 'No'],
    'B': ['Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'No']
})

# P(A = Yes and B = Yes)
joint_prob = len(data[(data['A'] == 'Yes') & (data['B'] == 'Yes')]) / len(data)
print(f"Joint Probability: {joint_prob:.2f}")

````



#### 2️⃣ Marginal Probability 

#### What is Marginal Probability?

> **Marginal probability** is the probability of a **single event happening**, regardless of other variables.

**Notation:**

$$
P(A) \quad \text{or} \quad P(B)
$$

It’s called **"marginal"** because in a probability table, it's often found in the **margins** (totals).

---

#### Real-Life Example

>  Suppose a hospital collects data on patients:

* A = “Has diabetes”
* B = “Is over 60 years old”

You might want to know:

* $P(A)$: Probability that a patient has diabetes (regardless of age)

---

#### Marginal from Joint Distribution

From a **joint probability table**, marginal probability is calculated by **summing rows or columns**.

#### Example Table:

|               | B = Yes  | B = No   | **Total (A)** |
| ------------- | -------- | -------- | ------------- |
| A = Yes       | 0.30     | 0.10     | **0.40**      |
| A = No        | 0.20     | 0.40     | **0.60**      |
| **Total (B)** | **0.50** | **0.50** | **1.00**      |

 **Marginal Probabilities:**

* $P(A = \text{Yes}) = 0.40$
* $P(B = \text{Yes}) = 0.50$

---

#### Marginal vs Joint vs Conditional

| Type        | Formula          | Meaning                                   |                                        |
| ----------- | ---------------- | ----------------------------------------- | -------------------------------------- |
| Joint       | $P(A \cap B)$    | Both A and B happen                       |                                        |
| Marginal    | $P(A)$ or $P(B)$ | Only A or only B, regardless of the other |                                        |
| Conditional | ( P(A            | B) )                                      | A happens **given** B already happened |

---




````python

import pandas as pd

data = pd.DataFrame({
    'Diabetes': ['Yes', 'No', 'Yes', 'Yes', 'No', 'No', 'Yes', 'No'],
    'Age_60plus': ['Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'No']
})

# P(Diabetes = Yes)
marginal_prob = len(data[data['Diabetes'] == 'Yes']) / len(data)
print(f"Marginal Probability of Diabetes: {marginal_prob:.2f}")

````


#### 3️⃣ **Law of Total Probability

---

####  What Is the Law of Total Probability?

The **Law of Total Probability** helps you calculate the **overall probability of an event** by **breaking it down into parts** — especially when you have conditional probabilities.

> It's used when:

* You know $P(A|B_i)$ (conditional probabilities)
* And $P(B_i)$ (probability of sub-events or partitions)

---

#### Formula:

If events $B_1, B_2, ..., B_n$ **partition the sample space**, and are **mutually exclusive and exhaustive**, then:

$$
P(A) = \sum_{i=1}^{n} P(A|B_i) \cdot P(B_i)
$$

---

#### Simple Real-Life Example

>  Suppose 70% of students are **undergraduates** (U), and 30% are **graduates** (G).
> The probability of **attending a seminar** is:

* 80% if Undergraduate → $P(S|U) = 0.8$
* 50% if Graduate → $P(S|G) = 0.5$

What is the **total probability** that a student attends the seminar?

---

#### Step-by-Step:

$$
P(S) = P(S|U) \cdot P(U) + P(S|G) \cdot P(G)
$$

$$
= (0.8 \times 0.7) + (0.5 \times 0.3)
$$

$$
= 0.56 + 0.15 = \boxed{0.71}
$$

✅ So, there's a **71% chance** that a randomly selected student attends the seminar.




````python

P_U = 0.7
P_G = 0.3
P_S_given_U = 0.8
P_S_given_G = 0.5

P_S = (P_S_given_U * P_U) + (P_S_given_G * P_G)
print(f"Total Probability of Attending Seminar: {P_S:.2f}")
````

#### Advanced Bayesian Inference 

---

#### 1. **Bayesian Inference Overview**

Bayesian inference is a framework for **updating beliefs about unknown parameters** using observed data, grounded in Bayes’ Theorem.

Given data $D$ and parameter(s) $\theta$, the goal is to find the **posterior distribution**:

$$
p(\theta | D) = \frac{p(D | \theta) \, p(\theta)}{p(D)}
$$

* $p(\theta)$ — Prior distribution: your belief about $\theta$ before seeing data
* $p(D | \theta)$ — Likelihood: probability of data given $\theta$
* $p(D) = \int p(D|\theta)p(\theta)d\theta$ — Evidence (marginal likelihood): normalizing constant
* $p(\theta | D)$ — Posterior distribution: updated belief about $\theta$ after seeing data

---

#### 2. **Key Concepts**

#### 2.1 Prior $p(\theta)$

* Encodes assumptions/beliefs about parameters before data
* Can be informative or non-informative (flat)
* Examples: Beta distribution for probabilities, Gaussian for continuous parameters

#### 2.2 Likelihood $p(D|\theta)$

* Model of how data is generated based on parameters
* Usually derived from probability distributions (e.g., Bernoulli, Normal)

#### 2.3 Evidence $p(D)$

* Integral over all parameter values:

$$
p(D) = \int p(D|\theta)p(\theta) d\theta
$$

* Ensures posterior is a valid probability distribution
* Often **difficult to compute analytically** in complex models

### 2.4 Posterior $p(\theta|D)$

* Combines prior and likelihood, updated with data
* Represents new knowledge about parameter(s) after observing data

---

#### 3. **Mathematical Derivation**

Starting from **Bayes’ Theorem**:

$$
p(\theta | D) = \frac{p(D | \theta) \, p(\theta)}{p(D)}
$$

The evidence is computed by:

$$
p(D) = \int p(D|\theta)p(\theta)d\theta
$$

---

#### 4. **Example: Bayesian Inference for Binomial Data**

---

#### Setup:

* You observe $k$ successes out of $n$ Bernoulli trials (coin flips)
* Likelihood:

$$
p(k | \theta) = \binom{n}{k} \theta^k (1 - \theta)^{n-k}
$$

* Prior: Beta distribution with parameters $\alpha, \beta$

$$
p(\theta) = \frac{\theta^{\alpha - 1} (1 - \theta)^{\beta - 1}}{B(\alpha, \beta)}
$$

where $B(\alpha, \beta)$ is the Beta function (normalizing constant).

---

#### Posterior distribution:

Because Beta is conjugate prior to Binomial likelihood, the posterior is also Beta:

$$
p(\theta | k) = \text{Beta}(\alpha + k, \beta + n - k)
$$

---

#### Interpretation:

* Prior belief: $\text{Beta}(\alpha, \beta)$
* Data observed: $k$ successes in $n$ trials
* Posterior belief: updated Beta with parameters $\alpha + k$, $\beta + n - k$

---

#### 5. **Computational Challenges**

* For complex models or continuous parameters, **exact posterior often unavailable**
* Use approximation methods:

  * **Markov Chain Monte Carlo (MCMC)**
  * **Variational Inference**
  * **Laplace Approximation**

---

## 6. **Bayesian vs Frequentist**

| Aspect                      | Bayesian                                       | Frequentist                           |
| --------------------------- | ---------------------------------------------- | ------------------------------------- |
| Parameters                  | Treated as random variables with distributions | Fixed but unknown                     |
| Interpretation              | Probability is degree of belief (subjective)   | Probability is long-run frequency     |
| Inference                   | Posterior distribution                         | Point estimates, confidence intervals |
| Incorporate prior knowledge | Yes                                            | No                                    |



````python
### Python Code (Bayesian Binomial Model)

from scipy.stats import beta
import matplotlib.pyplot as plt
import numpy as np

# Prior parameters
alpha_prior, beta_prior = 2, 2

# Data: k successes in n trials
k, n = 8, 10

# Posterior parameters
alpha_post = alpha_prior + k
beta_post = beta_prior + n - k

# Plot prior and posterior
x = np.linspace(0, 1, 100)

prior_pdf = beta.pdf(x, alpha_prior, beta_prior)
posterior_pdf = beta.pdf(x, alpha_post, beta_post)

plt.plot(x, prior_pdf, label='Prior Beta(2,2)')
plt.plot(x, posterior_pdf, label=f'Posterior Beta({alpha_post},{beta_post})')
plt.xlabel('Theta')
plt.ylabel('Density')
plt.title('Bayesian Updating for Binomial Data')
plt.legend()
plt.show()
````


#### Central Limit Theorem (CLT)

---

#### 1. **What is CLT?**

The **Central Limit Theorem** states that:

> Given a large enough sample size, the **sampling distribution of the sample mean** will be approximately **normally distributed** (Gaussian), regardless of the original population distribution.

---

#### 2. **Why is CLT important?**

* Allows us to use normal distribution to **approximate sampling distributions** of the mean.
* Makes statistical inference (like confidence intervals, hypothesis tests) possible **even if data is not normal**.
* Fundamental for many statistical methods and ML algorithms.

---

#### 3. **Formal Statement**

If you have independent, identically distributed (i.i.d.) random variables $X_1, X_2, ..., X_n$ each with:

* Mean $\mu$
* Variance $\sigma^2$

Then as $n \to \infty$:

$$
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}} \xrightarrow{d} N(0,1)
$$

Where:

* $\bar{X} = \frac{1}{n} \sum_{i=1}^n X_i$ is the sample mean
* $Z$ converges in distribution to a **standard normal distribution** $N(0,1)$

---

#### 4. **Intuition**

* Each individual sample mean varies due to sampling randomness.
* As you take **more samples**, the distribution of these sample means becomes **bell-shaped** and **normal**.
* The larger the sample size, the closer the distribution of $\bar{X}$ is to normal.

---

#### 5. **Key Points**

| Concept               | Explanation                                                 |
| --------------------- | ----------------------------------------------------------- |
| Sample Size           | Larger $n$ means better normal approximation                |
| Original Distribution | CLT holds **regardless of the original distribution** shape |
| Sample Mean           | Distribution of means becomes normal                        |
| Standard Error (SE)   | $\sigma / \sqrt{n}$, std. dev. of sample mean               |

---

#### 6. **Example**

Suppose you roll a fair six-sided die $n = 30$ times.

* Each die roll has mean $\mu = 3.5$, variance $\sigma^2 = \frac{35}{12} \approx 2.9167$
* By CLT, the distribution of the sample mean of these 30 rolls is approximately normal:

  $$
  \bar{X} \sim N\left(3.5, \frac{2.9167}{30}\right)
  $$

---

#### 7. **Practical Use**

You can approximate probabilities involving the sample mean by using the normal distribution — even if the original data is not normal.




````python

import numpy as np
import matplotlib.pyplot as plt

# Original distribution: uniform discrete (die rolls)
population = np.arange(1, 7)

# Parameters
mu = np.mean(population)
sigma = np.std(population)
n = 30  # sample size
num_samples = 10000

sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(population, n, replace=True)
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='skyblue')
plt.title('Sampling Distribution of Sample Mean (n=30)')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.show()

````

# 🔄 Markov Chains — Full Guide

---

## 1. **What is a Markov Chain?**

A **Markov Chain** is a stochastic process that undergoes transitions from one state to another on a state space. It has the **Markov property**:

> The probability of moving to the next state depends **only on the current state**, not on the sequence of states before it.

Mathematically:

$$
P(X_{n+1} = x | X_n = x_n, X_{n-1} = x_{n-1}, ..., X_0 = x_0) = P(X_{n+1} = x | X_n = x_n)
$$

---

## 2. **Components of a Markov Chain**

* **States**: Possible values the process can be in (finite or countable).
* **Transition probabilities**: Probability of moving from one state to another.
* **Initial distribution**: Probability distribution over states at time 0.

---

## 3. **Transition Matrix**

For a discrete Markov Chain with finite states $S = \{s_1, s_2, ..., s_m\}$, the transition probabilities can be represented as a matrix $P$:

$$
P = 
\begin{bmatrix}
p_{11} & p_{12} & \cdots & p_{1m} \\
p_{21} & p_{22} & \cdots & p_{2m} \\
\vdots & \vdots & \ddots & \vdots \\
p_{m1} & p_{m2} & \cdots & p_{mm} \\
\end{bmatrix}
$$

Where $p_{ij} = P(X_{t+1} = s_j | X_t = s_i)$ and

$$
\sum_{j=1}^m p_{ij} = 1, \quad \forall i
$$

---

## 4. **Types of Markov Chains**

* **Homogeneous**: Transition probabilities do not depend on time $t$.
* **Non-homogeneous**: Transition probabilities vary over time.

---

## 5. **Example: Weather Model**

Suppose weather states:

* $S = \{\text{Sunny}, \text{Rainy}\}$

Transition matrix:

$$
P = \begin{bmatrix}
0.8 & 0.2 \\
0.4 & 0.6 \\
\end{bmatrix}
$$

Interpretation:

* If today is sunny, 80% chance tomorrow is sunny, 20% rainy.
* If today is rainy, 40% chance tomorrow sunny, 60% rainy.

---

## 6. **Calculating Future State Probabilities**

If $\pi_t$ is a row vector representing distribution over states at time $t$, then:

$$
\pi_{t+1} = \pi_t \cdot P
$$

To find distribution after multiple steps:

$$
\pi_{t+k} = \pi_t \cdot P^k
$$

---

## 7. **Steady-State Distribution**

If the Markov chain is **irreducible** and **aperiodic**, it has a unique **stationary distribution** $\pi$ satisfying:

$$
\pi = \pi P
$$

This means if you multiply $\pi$ by $P$, you get $\pi$ again. It represents the long-term behavior.

---

## 8. **Markov Chains in NLP**

* **Hidden Markov Models (HMMs)** model sequences like parts of speech tagging.
* The Markov assumption simplifies modeling language as sequences with limited memory.

---

## 9. **Markov Chains in Reinforcement Learning (RL)**

* The environment state transitions are often modeled as Markov processes.
* RL algorithms use Markov Decision Processes (MDPs), which extend Markov chains by adding actions and rewards.

---

````python
import numpy as np

# Transition matrix for Sunny/Rainy example
P = np.array([[0.8, 0.2],
              [0.4, 0.6]])

# Initial state distribution: 100% Sunny
pi = np.array([1.0, 0.0])

# Predict weather for next 5 days
for day in range(5):
    pi = pi.dot(P)
    print(f"Day {day+1} distribution: Sunny={pi[0]:.3f}, Rainy={pi[1]:.3f}")
````
