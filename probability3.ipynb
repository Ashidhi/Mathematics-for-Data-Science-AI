

#  Discrete vs Continuous Probability Distributions

| Feature                        | **Discrete Distribution**                  | **Continuous Distribution**                                  |
| ------------------------------ | ------------------------------------------ | ------------------------------------------------------------ |
| **Definition**                 | Deals with **countable** outcomes          | Deals with **uncountable / infinite** outcomes               |
| **Random Variable**            | Takes **specific values** (e.g., 0, 1, 2…) | Takes **any value in an interval** (e.g., height = 165.3 cm) |
| **Probability Measured As**    | **Probability Mass Function (PMF)**        | **Probability Density Function (PDF)**                       |
| **Probability of exact value** | $P(X = x) > 0$                             | $P(X = x) = 0$ (exact value has zero probability)            |
| **Example Distributions**      | Bernoulli, Binomial, Poisson, Geometric    | Normal, Uniform, Exponential                                 |
| **Example Variables**          | # of emails, # of sales, coin tosses       | height, weight, time, temperature                            |
| **Area under curve**           | Sum of PMF values = 1                      | Total area under PDF curve = 1                               |



####  Key Differences Explained with Analogy:

| Concept                      | Discrete                                  | Continuous                                                                |
| ---------------------------- | ----------------------------------------- | ------------------------------------------------------------------------- |
| **Think of it like...**      | A **jar of marbles**: you can count them. | A **glass of water**: you can measure endlessly small portions.           |
| **Probability at one point** | $P(X=3) = 0.2$ is meaningful              | $P(X=2.0) = 0$, but area over an interval like $P(1.9 < X < 2.1)$ matters |
| **Total probability**        | $\sum P(X=x) = 1$                         | $\int_{-\infty}^{\infty} f(x)dx = 1$                                      |

---

####  Examples Comparison:

| Scenario                              | Discrete           | Continuous               |
| ------------------------------------- | ------------------ | ------------------------ |
| Coin toss                             | Number of heads    | —                        |
| Exam score out of 100 (whole numbers) | Discrete           | —                        |
| Exam score with decimals (e.g. 87.25) | —                  | Continuous               |
| Customer arrival count per hour       | Poisson (Discrete) | —                        |
| Time between customer arrivals        | —                  | Exponential (Continuous) |
| Heights of students                   | —                  | Normal (Continuous)      |

---

#### When to Use:

* Use **discrete** when dealing with **counting** (how many?).
* Use **continuous** when dealing with **measuring** (how much? how long?).

---

#### : From Discrete to Continuous

Some distributions have discrete and continuous analogs:

* **Binomial (Discrete)** ↔ **Normal (Continuous)** (Binomial approximates Normal when n is large)
* **Geometric (Discrete)** ↔ **Exponential (Continuous)** (both model waiting times)



```python
####  Visual Difference

#### Discrete:


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Discrete example: Binomial
x = np.arange(0, 11)
pmf = binom.pmf(x, n=10, p=0.5)
plt.stem(x, pmf, basefmt=" ")
plt.title("Discrete Distribution (Binomial)")
plt.xlabel("x")
plt.ylabel("P(X = x)")
plt.grid(True)
plt.show()

#### Continuous:

from scipy.stats import norm

# Continuous example: Normal
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, loc=0, scale=1)
plt.plot(x, pdf, color='green')
plt.title("Continuous Distribution (Normal)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```


#  Probability Distributions — Discrete

---

#### 1.  **Bernoulli Distribution**

#### ✅ Definition:

A **Bernoulli trial** is a random experiment with only **two possible outcomes**: success (1) or failure (0). The Bernoulli distribution models the outcome of a single trial.

#### ✅ Parameters:

* $p$: probability of success (0 ≤ p ≤ 1)
* $1 - p$: probability of failure

#### ✅ Probability Mass Function (PMF):

$$
P(X = x) = 
\begin{cases}
p & \text{if } x = 1 \\
1 - p & \text{if } x = 0
\end{cases}
$$

#### ✅ Mean and Variance:

* Mean: $\mu = p$
* Variance: $\sigma^2 = p(1 - p)$

#### ✅ Use Case:

* Tossing a biased coin (Head = 1, Tail = 0)
* Answering a true/false question correctly














```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

p = 0.6
data = bernoulli.rvs(p, size=1000)

# Plot
plt.hist(data, bins=[-0.5, 0.5, 1.5], rwidth=0.8, color='skyblue')
plt.xticks([0, 1])
plt.title("Bernoulli Distribution (p=0.6)")
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

#### 2.  **Binomial Distribution**

#### ✅ Definition:

The **Binomial Distribution** models the number of successes in **n independent Bernoulli trials**.

#### ✅ Parameters:

* $n$: number of trials
* $p$: probability of success

#### ✅ PMF:

$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
$$

#### ✅ Mean and Variance:

* Mean: $\mu = np$
* Variance: $\sigma^2 = np(1 - p)$

#### ✅ Use Case:

* Number of heads in 10 coin tosses
* Number of defective items in a batch


```python

from scipy.stats import binom

n, p = 10, 0.5
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

plt.bar(x, pmf, color='lightcoral')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

#### 3.  **Poisson Distribution**

#### ✅ Definition:

The **Poisson distribution** models the number of events occurring in a **fixed interval of time or space**, when events occur **independently** and at a **constant average rate** $\lambda$.

#### ✅ Parameters:

* $\lambda$: average number of events per interval

#### ✅ PMF:

$$
P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}
$$

#### ✅ Mean and Variance:

* Mean: $\mu = \lambda$
* Variance: $\sigma^2 = \lambda$

#### ✅ Use Case:

* Number of emails per hour
* Number of calls at a call center
* Number of accidents per day


```python

from scipy.stats import poisson

λ = 4
x = np.arange(0, 15)
pmf = poisson.pmf(x, λ)

plt.bar(x, pmf, color='mediumseagreen')
plt.title("Poisson Distribution (λ=4)")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

#### 4. **Geometric Distribution**

#### ✅ Definition:

The **Geometric distribution** models the number of **trials until the first success** in repeated, independent Bernoulli trials.

#### ✅ Parameters:

* $p$: probability of success in each trial

#### ✅ PMF:

$$
P(X = k) = (1 - p)^{k - 1} p
$$

Where $k \in \{1, 2, 3, ...\}$

#### ✅ Mean and Variance:

* Mean: $\mu = \frac{1}{p}$
* Variance: $\sigma^2 = \frac{1 - p}{p^2}$

#### ✅ Use Case:

* Number of coin tosses to get first head
* Number of attempts until you pass a test


```python

from scipy.stats import geom

p = 0.3
x = np.arange(1, 15)
pmf = geom.pmf(x, p)

plt.bar(x, pmf, color='orange')
plt.title("Geometric Distribution (p=0.3)")
plt.xlabel("Trial Number of First Success")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

#### Summary Table

| Distribution  | Models...                       | Parameters | Mean          | Variance            |
| ------------- | ------------------------------- | ---------- | ------------- | ------------------- |
| **Bernoulli** | 1 trial (0 or 1)                | $p$        | $p$           | $p(1 - p)$          |
| **Binomial**  | # of successes in $n$ trials    | $n, p$     | $np$          | $np(1 - p)$         |
| **Poisson**   | # of events in fixed interval   | $\lambda$  | $\lambda$     | $\lambda$           |
| **Geometric** | # of trials until first success | $p$        | $\frac{1}{p}$ | $\frac{1 - p}{p^2}$ |




## three most important Continuous Probability Distributions:


#### 1. Uniform Distribution (Continuous)

---

####  Definition:

A **Uniform Distribution** is one where **every value in a given range is equally likely**. This is called a **Continuous Uniform Distribution**.

> It’s like a perfectly fair spinner or selecting a random real number between 0 and 1.

---

####  PDF (Probability Density Function):

$$
f(x) = \frac{1}{b - a} \quad \text{for } a \le x \le b
$$

Where:

* $a$: lower bound
* $b$: upper bound

---

####  Properties:

* Mean: $\mu = \frac{a + b}{2}$
* Variance: $\sigma^2 = \frac{(b - a)^2}{12}$

---

####  Example Use Cases:

* Random number generator
* Equal chance between options (e.g., random time between 2–4pm)















```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

a, b = 2, 5
x = np.linspace(1, 6, 1000)
pdf = uniform.pdf(x, loc=a, scale=b-a)

plt.plot(x, pdf, color='orange')
plt.title("Uniform Distribution (a=2, b=5)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```


####  2. Normal Distribution (Gaussian)

---

####  Definition:

The **Normal Distribution** is the famous "bell curve" that models **real-world natural variations**.

$$
f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \, e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

Where:

* $\mu$: mean (center of the curve)
* $\sigma$: standard deviation (spread)

---

####  Properties:

* Symmetric around the mean
* Mean = Median = Mode
* Area under the curve = 1
* **68-95-99.7 Rule**:

  * 68% data within 1σ
  * 95% within 2σ
  * 99.7% within 3σ

---

#### Use Cases:

* Heights, weights, test scores
* Stock returns
* IQ scores

---

```python

from scipy.stats import norm

mu, sigma = 0, 1
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, mu, sigma)

plt.plot(x, pdf, color='blue')
plt.title("Normal Distribution (μ=0, σ=1)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

####  3. Exponential Distribution

---

#### Definition:

The **Exponential Distribution** models the time **between independent events** that happen at a **constant average rate** (Poisson process).

$$
f(x) = \lambda e^{-\lambda x} \quad \text{for } x \ge 0
$$

Where:

* $\lambda$: rate parameter (events per unit time)
* $1/\lambda$: average waiting time

---

####  Properties:

* Mean: $\mu = 1/\lambda$
* Variance: $\sigma^2 = 1/\lambda^2$
* Skewed right (not symmetric)
* Memoryless: $P(X > s + t \mid X > s) = P(X > t)$

---

####  Use Cases:

* Time between arrivals at a bus stop
* Time until failure of a machine
* Time between phone calls

---

```python

from scipy.stats import expon

λ = 1  # rate
x = np.linspace(0, 10, 1000)
pdf = expon.pdf(x, scale=1/λ)

plt.plot(x, pdf, color='green')
plt.title("Exponential Distribution (λ=1)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

| Uniform      | Flat line    | $a, b$        | Random real value between 2 and 5 |
| Normal       | Bell curve   | $\mu, \sigma$ | Human height, test scores         |

#### What is the Mean (Expected Value)?

For a discrete random variable $X$ that can take values $x_1, x_2, ..., x_n$ with probabilities $P(x_1), P(x_2), ..., P(x_n)$:

####  Formula:

$$
\mu = E(X) = \sum_{i=1}^{n} x_i \cdot P(x_i)
$$

> Multiply each outcome by its probability, then sum them all.

---

#### Manual Example

| Outcome $x_i$ | Probability $P(x_i)$ |
| ------------- | -------------------- |
| 0             | 0.1                  |
| 1             | 0.3                  |
| 2             | 0.4                  |
| 3             | 0.2                  |

#### Step-by-step:

$$
E(X) = (0)(0.1) + (1)(0.3) + (2)(0.4) + (3)(0.2) = 0 + 0.3 + 0.8 + 0.6 = \boxed{1.7}
$$



#### 🧠 Summary:

* The **mean of a discrete distribution** gives the **long-run average outcome**.
* It is also called the **expected value**, denoted as $E(X)$.
* You need:

  * All possible outcomes $x_i$
  * Their respective probabilities $P(x_i)$
* The probabilities **must sum to 1**.



```python

import numpy as np

# Values and probabilities
x = np.array([0, 1, 2, 3])
p = np.array([0.1, 0.3, 0.4, 0.2])

# Check: sum of probabilities must be 1
assert np.isclose(p.sum(), 1.0), "Probabilities must sum to 1"

# Mean (Expected value)
mean = np.sum(x * p)
print(f"Mean (Expected Value) of X = {mean}")
```


####  1. **Expected Value (Mean)**

It tells you the **long-run average** of a random variable after many repetitions.

####  Formula for **Discrete Random Variable**:

$$
E[X] = \sum x \cdot P(x)
$$

* $x$: value the variable can take
* $P(x)$: probability of that value
* This is a **weighted average**

####  Example (Dice Roll):

Roll a fair 6-sided die:

$$
E[X] = 1\cdot\frac{1}{6} + 2\cdot\frac{1}{6} + \dots + 6\cdot\frac{1}{6} = \frac{21}{6} = 3.5
$$

---

#### 2. **Variance (Var\[X])**

Variance tells you **how spread out** the values are around the expected value.

####  Formula:

$$
Var(X) = \sum P(x) \cdot (x - E[X])^2
$$

It’s the **average of squared differences** from the mean.

---

#### 3. **Standard Deviation (SD or σ)**

Just the **square root of variance** — brings it back to the original units.

$$
\sigma = \sqrt{Var(X)}
$$

---

#### Full Example:

Let’s say a variable $X$ takes values 1, 2, and 3 with:

* $P(1) = 0.2$
* $P(2) = 0.5$
* $P(3) = 0.3$

#### Step 1: **Expected Value**

$$
E[X] = 1(0.2) + 2(0.5) + 3(0.3) = 0.2 + 1.0 + 0.9 = 2.1
$$

#### Step 2: **Variance**

$$
Var(X) = 0.2(1 - 2.1)^2 + 0.5(2 - 2.1)^2 + 0.3(3 - 2.1)^2
$$

$$
= 0.2(1.21) + 0.5(0.01) + 0.3(0.81) = 0.242 + 0.005 + 0.243 = 0.49
$$

#### Step 3: **Standard Deviation**

$$
\sigma = \sqrt{0.49} = 0.7
$$


```python

import numpy as np

x = np.array([1, 2, 3])          # values
p = np.array([0.2, 0.5, 0.3])    # probabilities

# Expected Value
E_X = np.sum(x * p)

# Variance
Var_X = np.sum(p * (x - E_X)**2)

# Standard Deviation
Std_X = np.sqrt(Var_X)

print(f"Expected Value: {E_X}")
print(f"Variance: {Var_X}")
print(f"Standard Deviation: {Std_X}")


```
